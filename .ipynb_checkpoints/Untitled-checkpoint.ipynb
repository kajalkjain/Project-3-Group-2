```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# File to Load 
file_clean_data = Path("data/cleaned_healthcare_dataset.csv.csv")

# Read cleaned_healthcare_dataset and store into Pandas DataFrames
df = pd.read_csv(file_clean_data)
```

```python
# What trends can be observed in test results over time?
# Visualization: Line plot.(Jess)

# Sample data: months and corresponding test scores
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
scores = [65, 59, 80, 81, 56, 55, 40, 70, 90, 85, 78, 95]

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(months, scores, marker='o')

# Customize the plot
plt.title('Test Results Trend Over Time', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Test Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Add value labels on the data points
for i, score in enumerate(scores):
    plt.text(i, score, str(score), ha='center', va='bottom')

# Calculate and plot trend line
z = np.polyfit(range(len(months)), scores, 1)
p = np.poly1d(z)
plt.plot(months, p(range(len(months))), "r--", alpha=0.8, label='Trend Line')

plt.legend()
plt.tight_layout()

# Show the plot
plt.show()

# Calculate some basic statistics
average_score = np.mean(scores)
min_score = np.min(scores)
max_score = np.max(scores)

print(f"Average score: {average_score:.2f}")
print(f"Minimum score: {min_score}")
print(f"Maximum score: {max_score}")
print(f"Score range: {max_score - min_score}")
```
