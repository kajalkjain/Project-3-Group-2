```python
# How do different blood types correlate with medical conditions?
#Visualization: Heatmap or stacked bar chart.(Kajal)

# Create a pivot table to count the occurrences of each Medical Condition by Blood Type
pivot_table = df.pivot_table(index='Blood Type', columns='Medical Condition', aggfunc='size', fill_value=0)

#The pivot_table() function creates a matrix where rows are blood types and columns 
#are medical conditions, with each cell showing the count of how often that blood type has that condition.

plt.figure(figsize=(10, 6))

# Create the heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", linewidths=0.5)

#sns.heatmap() visualizes this table as a heatmap. The annot=True argument shows the count values in the cells.

# Customize the heatmap
plt.title('Correlation between Medical Conditions and Blood Types')
plt.ylabel('Blood Type')
plt.xlabel('Medical Condition')

# Show the plot
plt.tight_layout()
plt.show()
```

The heatmap shows how each blood type correlates with specific medical conditions. Darker cells represent a higher occurrence, 
while lighter cells represent a lower occurrence.


```python
#- What are the most common medical conditions per hospital?
#Visualization: Bar chart or heatmap.(Kajal)

# Group the data by Hospital and Medical Condition and count occurrences
condition_count = df.groupby(['Hospital', 'Medical Condition']).size().unstack(fill_value=0)

# Plotting the stacked bar chart using seaborn or matplotlib
condition_count.plot(kind='bar', stacked=True, figsize=(12, 8), cmap="tab20")

# Customize the chart
plt.title('Most Common Medical Conditions per Hospital')
plt.ylabel('Number of Cases')
plt.xlabel('Hospital')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

The chart shows the most common medical conditions for each hospital. The taller the bar, the more cases of a particular 
medical condition in that hospital.
Different medical conditions by distinguished by color, as each segment of the bar represents a different condition.
